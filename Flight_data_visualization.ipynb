```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("ggplot")  # daha güzel grafik görünümü
```

```python
df["name"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Airlines by Number of Flights")
plt.xlabel("Airline")
plt.ylabel("Flight Count")
plt.xticks(rotation=45)
plt.show()
```

```python
airline_delay = df.groupby("name")["arr_delay"].mean().sort_values(ascending=False)

airline_delay.head(10).plot(kind="bar")

plt.title("Average Arrival Delay by Airline")
plt.xlabel("Airline")
plt.ylabel("Average Arrival Delay (minutes)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
origin_delay = df.groupby("origin")["arr_delay"].mean()

origin_delay.plot(kind="bar")

plt.title("Average Arrival Delay by Origin Airport")
plt.xlabel("Origin Airport")
plt.ylabel("Average Arrival Delay (minutes)")
plt.tight_layout()
plt.show()
```

```python
df["month"].value_counts().sort_index().plot(kind="bar")

plt.title("Number of Flights by Month")
plt.xlabel("Month")
plt.ylabel("Number of Flights")
plt.tight_layout()
plt.show()
```
